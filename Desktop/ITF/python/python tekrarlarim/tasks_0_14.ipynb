```python
"""
Task-13: Find out the most frequent number and its frequency.
"""
nums = [1,30,4,5,4,30,0,9,30,30]
fre = max(nums, key=nums.count)
print(fre)
print(f'the most freq. number is {fre} and it was {nums.count(fre)} times repeated.')
```

```python
from collections import Counter
numbers = [1,3,4,5,4,3,0,3,6,3]
freq_num = Counter(numbers)
print(f"the most frequent number is {freq_num.most_common()[0][0]} and it was {freq_num.most_common()[0][1]} times repeated.")
```

```python
from statistics import mode
numbers = [1,3,4,5,4,3,0,3,6,3]
freq_num = mode(numbers)
print(f"the most frequent number is {freq_num} and it was {numbers.count(freq_num)} times repeated.")
```
